```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to load
customers_table_file = "Resources/customers_table.csv"
review_id_file = "Resources/review_id_table.csv"
vine_table_file = "Resources/vine_table.csv"
vine_customer_file = "Resources/vine_customers.csv"

# Read file into dataframe
customers_table_df = pd.read_csv(customers_table_file)
review_id_df = pd.read_csv(review_id_file)
vine_table_df = pd.read_csv(vine_table_file)
vine_customer_df = pd.read_csv(vine_customer_file)
```

```python
# New dataframe where total_votes is more than 20 votes
vine_20_votes = vine_table_df.loc[vine_table_df["total_votes"] >= 20]
vine_20_votes.head(10)
```

```python
# Filter vine_20_votes where helpful_votes/total_votes is greater than 50%
vine_helpful_over_total = vine_20_votes.loc[(vine_20_votes["helpful_votes"] / (vine_20_votes["total_votes"]) >= 0.5)]
vine_helpful_over_total.head(10)
```

```python
# Dataframe Vine program (paid), vine == 'Y'
vine_paid = vine_helpful_over_total.loc[vine_helpful_over_total["vine"] == "Y"]
vine_paid.head(10)
```

```python
# Dataframe Vine program (unpaid), vine == 'N'
vine_unpaid = vine_helpful_over_total.loc[vine_helpful_over_total["vine"] == "N"]
vine_unpaid.head(10)
```

```python
# Total vine and non-vine reviews
vine_count = len(vine_paid)
non_Vine_count = len(vine_unpaid)
print(f"Total number of vine reviews: {vine_count} and total number of non-vine reviews {non_Vine_count}.")
```

```python
# Total number of reviews
total_reviews = len(vine_helpful_over_total)
total_reviews
```

```python
# The number of 5-star reviews
total_5_star = len(vine_helpful_over_total.loc[vine_helpful_over_total["star_rating"] == 5])
total_5_star
```

```python
# Total number of paid (vine) 5 star reviews
paid_5_star = len(vine_paid.loc[vine_paid["star_rating"] == 5])
paid_5_star
```

```python
# Total number of unpaid 5 star reviews
unpaid_5_star = len(vine_unpaid.loc[vine_unpaid["star_rating"] == 5])
unpaid_5_star
```

```python
# The percentage of 5-star reviews for paid
percent_paid_5_star = (paid_5_star / total_5_star) * 100
round(percent_paid_5_star, 2)
```

```python
# The percentage of 5-star reviews for unpaid
percent_unpaid_5_star = (unpaid_5_star / total_5_star) * 100
round(percent_unpaid_5_star, 2)
```

```python
vine_review = pd.DataFrame([{"Total Reviews": total_reviews, "Total 5 Star": total_5_star, "Paid 5 star": paid_5_star, "Unpaid 5 star": unpaid_5_star, "Percent Paid 5 Star": percent_paid_5_star, "Percent Unpaid 5 Star": percent_unpaid_5_star}])
vine_review.style.hide_index()
```
